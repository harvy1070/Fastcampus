## 5-6. ë°ì´í„° ë°±ì—…í•˜ê¸°

> - ë°ì´í„° ë² ì´ìŠ¤ë¥¼ ë°±ì—…í•  ë•Œ ì‚¬ìš©í•˜ëŠ” ëª¨ë“ˆ
> - `.sql` íŒŒì¼ í™•ì¥ìë¡œ **í…Œì´ë¸”ì„ ë‹¤ì‹œ ë³µì›**í•  ìˆ˜ ìˆëŠ” ì¿¼ë¦¬ë¬¸ì„ ì €ì¥

```python
import sqlite3
```

```python
# ì—°ìŠµìš© DB ì—°ê²°
conn = sqlite3.connect("test.db")
# ì»¤ì„œ ìƒì„±
c = conn.cursor()
```

```python
# ì¶”ê°€í•  ìƒí’ˆ ë¦¬ìŠ¤íŠ¸
product_list = [[1, 'ëª¨ì', 15000],
                [2, 'ì½”íŠ¸', 200000],
                [3, 'í‹°ì…”ì¸ ', 20000],
                [4, 'ë¸”ë¼ìš°ìŠ¤', 55000],
                [5, 'ê°€ë””ê±´', 45000],
                [6, 'ì²­ë°”ì§€', 50000],
                [7, 'êµ¬ë‘', 150000],
                [8, 'ê°€ë°©', 170000]]
```

```python
# ë°ì´í„° ì—¬ëŸ¬ì¤„ ì‚½ì…
c.executemany("INSERT OR REPLACE INTO test(ID, PRODUCT_NAME, PRICE) VALUES(?,?,?)", product_list)
conn.commit()
```

```python
# iterdump ë‚´ìš© í™•ì¸
for line in conn.iterdump():
    print(line)
```

```python
# ë°ì´í„° ë² ì´ìŠ¤ ë°±ì—… íŒŒì¼ ìƒì„±
with conn:
    with open('backup.sql', 'w') as f:
        for line in conn.iterdump():
            f.write('%s\n' % line)
        print('Completed.')
```

```python
# í…Œì´ë¸” ì‚­ì œ
c.execute("DROP TABLE test")
conn.commit()
```

```python
# ë°±ì—… SQL íŒŒì¼ ë¡œë”©
with open('backup.sql', 'r') as sql_file:
    sql_script = sql_file.read()
```

```python
# SQL ìŠ¤í¬ë¦½íŠ¸ ì‹¤í–‰
c.executescript(sql_script)
conn.commit()
```
